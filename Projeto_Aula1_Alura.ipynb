# #Desafio 7

```python
#Desafio 7
import matplotlib.pyplot as plt
%matplotlib inline

contagem_tempo = dados['tempo'].value_counts().values
grupo_tempo = dados['tempo'].value_counts().index

plt.figure(figsize = (5, 3))
plt.bar(grupo_tempo, contagem_tempo, ec = "k", alpha = .8, color = "purple")
plt.xlabel('Tempo', fontsize = 12)
plt.ylabel('n° de amostras', fontsize = 12)
plt.title('Ordenação de amostras por Tempo', fontsize = 10)
```
